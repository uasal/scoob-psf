```python
arrs1 = np.random.randn(100,10,10)
arrs2 = np.random.randn(100,10,10)

```

```python

def create_anim(arrs1,arrs2,):
    numframes = arrs1.shape[0]
    print(numframes)

    fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,4),dpi=150)
    
    im1 = ax[0].imshow(arrs1[0,:,:],)
    im2 = ax[1].imshow(arrs2[0,:,:],)
    im1_title = ax[0].set_title(f'Frame {1:d}', fontsize = 18)

    im2_title = ax[1].set_title(f'Frame {1:d}', fontsize = 18)

    def animate(i):
        im1.set_data(arrs1[i,:,:])
        im1_title.set_text(f'Frame {1:d}')

        im2.set_data(arrs2[i,:,:])
        im2_title.set_text(f'Frame {1:d}')

    anim = matplotlib.animation.FuncAnimation(fig, animate, frames=numframes)
    return anim
```

```python
create_anim(arrs1, arrs2)
```
